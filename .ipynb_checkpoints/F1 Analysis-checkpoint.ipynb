```python
def complists(l1,l2):
    l1 = l1.sort_values(3,ignore_index=True)
    l2 = l2.sort_values(3,ignore_index=True)
    d1 = l1.iloc[:,2]
    d2 = l2.iloc[:,2]
    
    order = []
    for i,d in l1.iterrows():
        if d[2] in list(d2):
            order.append([d[0],d[1],d[2],d2[d2==d[2]].index.values[0]])
    
    order = pd.DataFrame(order)
    order = np.array(order)
    return order

def mergeSort(arr,dic):
    if len(arr) > 1:
        temp = []
  
         # Finding the mid of the array
        mid = len(arr)//2
  
        # Dividing the array elements
        L = arr[:mid]
  
        # into 2 halves
        R = arr[mid:]
  
        # Sorting the first half
        L = mergeSort(L,dic)
  
        # Sorting the second half
        R = mergeSort(R,dic)
  
        i = 0
        j = 0
  
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i][3] < R[j][3]:
                temp.append(list(L[i]))
                i += 1
            else:
                temp.append(list(R[j]))
                vals = dic[R[j][2]]
                dic[R[j][2]] = [vals[0]+len(L)-i,vals[1]]
                for x in range(i,len(L)):
                    vals = dic[L[x][2]]
                    dic[L[x][2]] = [vals[0],vals[1]+1]
                j += 1
  
        # Checking if any element was left
        while i < len(L):
            temp.append(list(L[i]))
            i += 1
  
        while j < len(R):
            temp.append(list(R[j]))
            j += 1
        
        return temp
    return arr
```

```python
def overtakes(filename,dic,pits,rm=True):
    if os.path.exists(filename):
        f = pd.read_csv(filename)
        raceId = f['raceId']
        if int(raceId[0]) > 840:
            driverId = f['driverId']
            racers = f.iloc[:,3]
            prev = f.iloc[:,7]
            prev = pd.concat([raceId,driverId,racers,prev],keys=[0,1,2,3],axis=1)
            stops = pits.loc[pits['raceId'] == raceId[0]]
            for c in f.columns[8:]:
                times = f[c]
                times = pd.concat([raceId,driverId,racers,times],keys=[0,1,2,3],axis=1)
                times.dropna(subset = [3], inplace=True)
                if rm:
                    if int(c)-1 in list(stops['lap']):
                        rows = stops[stops['lap'] == int(c)-1]
                        for i,r in rows.iterrows():
                            times = times[times[1] != r[1]]
                            prev = prev[prev[1] != r[1]]
                changes = complists(prev,times)
                sort = mergeSort(changes,dic)
                prev = f[c]
                prev = pd.concat([raceId,driverId,racers,prev],keys=[0,1,2,3],axis=1)
```

```python
dic = {}
for r in drivers['driverRef']:
    dic[r] = [0,0]
print(1)
for d in dirs:
    overtakes('./races/{}/total_times.csv'.format(d),dic,pits)

for k in list(dic.keys()):
    if dic[k] == [0,0]:
        dic.pop(k)
```
